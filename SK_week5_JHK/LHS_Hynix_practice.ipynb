행렬로하여 코드를 짤수도있지만 직관적으로 보이게하기위해서 하나하나 작성하였다.

### step1  
먼저 필요한 라이브러리를 import해주며 변수의 갯수와 level의 수를 정의해줍니다.(level은 변수마다 같아야 사용가능)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import csv    

m=3; 
x=100; 
```

### step2
각 변수값의 최대, 최소값을 정의한다.  

```python
max_variable=[100,80,100];
min_variable=[50,20,10];
```

### step3
정해진 변수의 값을 레벨의 수만큼 random하지만 겹치지않게 출력해줍니다. np.random.permutation 이용.
하지만 위함수를 그냥사용하면 0부터 x-1까지만 출력이된다.

```python
x1,x2,x3 = [np.random.permutation(np.linspace(min_variable[i],max_variable[i],100)).round(2) 
            for i in range(len(max_variable))]
```

```python
len(x1),len(x2),len(x3)
```

### step6
LHS가 잘되었는지 확인해본다. 2d그래프와 3d그래프로 확인

```python
plt.figure(1)
plt.plot(x1,x2,'.')
plt.grid(True)
plt.xlabel('x1 variable')
plt.ylabel('x2 variable')
plt.title('LHS-2variables')

plt.figure(2)
plt.plot(x1,x3,'.')
plt.grid(True)
plt.xlabel('x1 variable')
plt.ylabel('x3 variable')
plt.title('LHS-2variables')

plt.figure(3)
plt.plot(x2,x3,'.')
plt.grid(True)
plt.xlabel('x2 variable')
plt.ylabel('x3 variable')
plt.title('LHS-2variables')
```

3D 그래프 확인.

```python
fig=plt.figure(2)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, x3,s= 20, alpha=0.5)
plt.xlabel('x1 variable')
plt.ylabel('x2 variable')
plt.ylabel('x3 variable')
plt.title('LHS-3variables')
```

### step 7
LHS로구해준 값들을 csv파일로 저장한다. csv함수는 column으로 저장하는것을 지원하지 않는다.

```python
f = open('LHS_Hynix.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(f)
wr.writerow(x1)
wr.writerow(x2)
wr.writerow(x3)
f.close()
```
